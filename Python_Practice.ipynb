```
# Reverse digits using while loop
num=int(input("Enter a no. "))
absnum=abs(num)
rev=absnum%10
absnum=absnum//10
while absnum>0:
  r=absnum%10
  absnum=absnum//10
  rev=rev*10+r
if (num>=0):
    print(rev)
else:
    print(rev-2*rev)
```

```
#Check for palindrome using while loop
#A palindrome is a no. that is same even when it's digits are reversed
num=int(input("Enter a no. "))
absn=abs(num)
rev=absn % 10
absn=absn // 10
while (absn > 0):
  r=absn % 10
  absn=absn // 10
  rev = rev*10+r
if (num == rev):
  print("It is a Palindrome")
else:
  print("Not a Palindrome")
```

```
#Find all prime no.s less than the entered no. using for loop
num=int(input('Enter a no. '))
if num>2:
  print(2,end=' ')
for i in range(3,num):
  flag=False
  for j in range(2,i):
    if i%j==0:
      flag=False
      break
    else:
      flag=True
  if flag:
    print(i,end=' ')
```

```
# Find the total profit/loss of each trader working in a trading firm. Keep accepting the trade amount until it is 0
emp_id=input('Enter the emp no.')
trade=int(input('Enter the trade amount '))
profit_loss=0
while (trade !=0):
  profit_loss=profit_loss +trade
  trade=int(input('Enter the trade amount '))
print(f'Profit/Loss for employee ID {emp_id} is {profit_loss}')
```

```
# Break- It breaks the loop at the given character
# For eg. To enter the username from the email id
email=input('Enter your email ID: ')
for c in email:
  if (c=='@'):
    break
  print(c, end='')
```

```
#Continue- If we want to print all characters except the given character
#Eg. - Enter all characters of the email ID except @ sign
email=input('Enter your email ID: ')
for c in email:
  if (c=='@'):
    continue
  print(c, end='')
```
